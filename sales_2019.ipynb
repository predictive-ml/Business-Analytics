## DATASET OF INTEREST

In order to meet the objective, we shall extract the transactions of interest. These are transactions where the quantity is positive (>0) and the amount is positive (>0). This implies that item(s) were sold and revenue was generated for the business.  

```python
# rows with positive values in Quantity column and Amount column
# df_clean consits of transactions that bring in revenue
df_clean = df_copy[(df_copy.amount>0) & (df_copy.quantity>0)]
```

```python
#create new column 'month_name' that registered year sale transaction was held(helps with analysis)
df_clean['month_name'] = df_clean.date.dt.month_name()
```

```python
#create new column 'month_number' that registered year sale transaction was held(helps with analysis)
df_clean['month_number'] = df_clean.date.dt.month
```

**Test**

```python
df_clean.head() #preview five first rows
```

```python
#Total revenue generated for January - December 2019
sum(amount for amount in df_clean.amount) 
```

```python
#Total quantity sales for January - December 2019
sum(qty for qty in df_clean.quantity) 
```

```python
#Total quantity of items sold and corresponding revenue generated for 2019
qty_revenue_2019= df_clean.groupby(
   ['month_name']
).agg(
    {
         'amount':sum,    # Sum revenue per customer
         'quantity': sum  # get the sum of items sold per year
         
    }
)

qty_revenue_2019.sort_values(by=['amount'], inplace=True, ascending=False) 
```

```python
qty_revenue_2019
```

```python
#  PIVOT TABLE
qty_revenue_table = pd.pivot_table(df_clean, index="customer_category",columns='month_name',
                              values =["amount","quantity"],aggfunc=sum, margins=True)

#sort the pivot table
qty_revenue_table.sort_values(by=('amount','All'), ascending=False, inplace= True)
```
